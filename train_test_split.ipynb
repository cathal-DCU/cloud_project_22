```python
# https://towardsdatascience.com/step-by-step-tutorial-pyspark-sentiment-analysis-on-google-dataproc-fef9bef46468
# import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# set the names for each column
cols = ['sentiment','id','date','query_string','user','text']
file_to_open = "gs://cloud-project-bucket-3/Sentiment_Analysis_Dataset.csv"
df = pd.read_csv(file_to_open, encoding = 'ISO-8859-1')
print(df.dtypes)
```

```python
"""
	# set the random seed and split train and test with 99 to 1 ratio
	np.random.seed(777)
	msk = np.random.rand(len(df)) < 0.99
	train = df[msk].reset_index(drop=True)
	test = df[~msk].reset_index(drop=True)
    """

def main():
    # read training data with ISO-8859-1 encoding and column names set above
    df = pd.read_csv(file_to_open, encoding = 'ISO-8859-1',names=cols)
    # shuffle the data
    df = df.sample(frac=1).reset_index(drop=True)
    
    # split into train and test sets
    train, test = train_test_split(df, test_size=0.2)
    
    # save both train and test as CSV files
    train.to_csv('gs://cloud-project-bucket-3/pyspark_sa_train_data.csv')
    test.to_csv('gs://cloud-project-bucket-3/pyspark_sa_test_data.csv')

if __name__=="__main__":
	main()
```
